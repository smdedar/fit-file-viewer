## activity

```python
def get_activity_data(fitfile):
    df = pd.DataFrame( (
        (
            activity.get_value("timestamp"),
            activity.get_value("total_timer_time"),
            activity.get_value("local_timestamp"),
            activity.get_value("num_sessions"),
            activity.get_value("type"),
            activity.get_value("event"),
            activity.get_value("event_type"),
        )
        for activity in fitfile.get_messages("activity")
    ),
    columns=[
        "timestamp",
        "total_timer_time",
        "local_timestamp",
        "num_sessions",
        "type",
        "event",
        "event_type",
    ],
    )
    return df.to_json(orient="records")


print(get_activity_data(fitfile))
   
```

```python
def get_device_info(fitfile):
    df = pd.DataFrame(
        (
            (
                device_info.get_value("timestamp"),
                device_info.get_value("serial_number"),
                device_info.get_value("manufacturer"),
                device_info.get_value("product"),
                device_info.get_value("software_version"),
                device_info.get_value("device_index"),
                device_info.get_value("device_type"),
            )
            for device_info in fitfile.get_messages("device_info")
        ),
        columns=[
            "timestamp",
            "serial_number",
            "manufacturer",
            "product",
            "software_version",
            "device_index",
            "device_type",
        ],
    )
    return df.to_json(orient="records")

print(get_device_info(fitfile))

```

```python
device_info_index = []
for device_info in fitfile.get_messages("device_info"):
    for data in device_info:
        if data.name not in device_info_index:
            print(data.name)
            device_info_index.append(data.name)
```
